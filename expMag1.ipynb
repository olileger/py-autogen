```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

import os
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_agentchat.agents import AssistantAgent
from autogen_agentchat.teams import MagenticOneGroupChat
from autogen_agentchat.ui import Console
```

```python
async def main() -> None:

    gpt4o = OpenAIChatCompletionClient(model="gpt-4o", api_key=os.environ["API_KEY"])
    assistant = AssistantAgent("Assistant", model_client=gpt4o)
    team = MagenticOneGroupChat([assistant], model_client=gpt4o)

    await Console(team.run_stream(task="Provide a different proof for Fermat's Last Theorem"))
```

```python
asyncio.run(main())
```
